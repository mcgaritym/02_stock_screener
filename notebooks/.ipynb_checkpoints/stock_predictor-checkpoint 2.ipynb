# Stock Predictor

## 0. Introduction:

The purpose of this project is to evaluate stock prices and employ various machine learning methods to try and predict their behavior in the future. The motivation for this project is due to personal interest (investing), and the fact that stocks provide a wealth of historical data to include in a machine learning problem. Also, I am interested in determining the effect of the COVID-19 pandemic on stocks in general. 

This project will try and answer these COVID-19 related questions:

- Which stocks have risen (or fallen) the most during COVID-19?
- Which stocks rise and fall together (clustering) before vs. during COVID-19?
- Which stocks are correlated to COVID-19 virus/lockdown sentiment?
- How have stock fundamentals/financials changed before vs. during COVID-19?

Also, the project will try and answer these general questions:

- Can we predict future stock prices (supervised regression) using forecasting techniques?
- Can we predict future stock prices (supervised regression) using sentiment data or company financial data?
- Can we predict if stock prices will rise or fall (supervised classification) using sentiment data or company financial data?
- Practice: Set up a ETL pipeline and display stock pick recommendation on dashboard

### 3.1 Stock Clustering Before/During COVID-19

```python
# create percent change columns of closing price
close_prices_rolling['AAPL_%'] = close_prices_rolling['AAPL_mean'].pct_change()
close_prices_rolling['FB_%'] = close_prices_rolling['FB_mean'].pct_change()
close_prices_rolling['AMZN_%'] = close_prices_rolling['AMZN_mean'].pct_change()
close_prices_rolling['NFLX_%'] = close_prices_rolling['NFLX_mean'].pct_change()
close_prices_rolling['GOOGL_%'] = close_prices_rolling['GOOGL_mean'].pct_change()
close_prices_rolling['MSFT_%'] = close_prices_rolling['MSFT_mean'].pct_change()
close_prices_rolling['CCL_%'] = close_prices_rolling['CCL_mean'].pct_change()
close_prices_rolling['AAL_%'] = close_prices_rolling['AAL_mean'].pct_change()
close_prices_rolling['LUV_%'] = close_prices_rolling['LUV_mean'].pct_change()

# create labels
labels = ['AAPL', 'FB', 'AMZN', 'NFLX', 'GOOGL', 'MSFT', 'CCL', 'AAL', 'LUV']

# create numpy array of stock price changes
movements = close_prices_rolling[['AAPL_%', 'FB_%','AMZN_%',
                                                        'NFLX_%',
                                                        'GOOGL_%',
                                                        'MSFT_%',
                                                        'CCL_%',
                                                        'AAL_%',
                                                        'LUV_%']].dropna(axis=0).to_numpy().transpose()

# normalized_movements = normalize(close_prices_rolling[['AAPL_%', 
#                                                        'FB_%',
#                                                        'AMZN_%',
#                                                        'NFLX_%',
#                                                        'GOOGL_%',
#                                                        'MSFT_%']])

# calculate the linkage: mergings
mergings = linkage(movements, method='complete')

# plot the dendrogram
plt.figure(dpi=250)
dendrogram(mergings, labels=labels, leaf_rotation=90, leaf_font_size=6)
plt.show()
```
